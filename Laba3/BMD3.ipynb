```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.distplot(tips['total_bill'], bins = 20)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.pairplot(tips, hue = 'day')
```

```python
import numpy as np
sns.barplot(x = 'sex', y = 'total_bill', data = tips, estimator = np.std, hue = 'day')
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', palette = 'rainbow')
```

```python
sns.stripplot(x = 'sex', y = 'total_bill', data = tips, hue = 'day', dodge = True)
```

```python
sns.heatmap(tips.corr(), annot = True)
```

```python
graf = sns.FacetGrid(tips, col = "time", row = "smoker")
```

```python
from matplotlib import pyplot as plt
graf = sns.FacetGrid(tips, col = "time", row = "smoker")
graf = graf.map(plt.hist, "total_bill")
```
