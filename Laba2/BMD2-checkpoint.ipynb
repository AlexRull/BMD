```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('Вісь X')
plt.ylabel('Вісь Y')
plt.title('Парабола')
```

```python
plt.subplot(1, 3, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 3, 2)
plt.plot(x, y, 'g')

plt.subplot(1, 3, 3)
plt.plot(x, y, 'b')
```

```python
figur = plt.figure()

axes = figur.add_axes([0, 0, 1, 1])

axes.plot(x, y)
axes.set_xlabel('Вісь X')
axes.set_ylabel('Вісь Y')
axes.set_title('Парабола')
```

```python
figur = plt.figure()

axes_1 = figur.add_axes([0, 0, 1, 1])
axes_2 = figur.add_axes([0.1, 0.5, 0.4, 0.3])
axes_3 = figur.add_axes([0.7, 0.2, 0.2, 0.2])

axes_1.plot(x, y, 'r')
axes_1.set_xlabel('Вісь X')
axes_1.set_ylabel('Вісь Y')
axes_1.set_title('Парабола')

axes_2.plot(x, y, 'g')
axes_2.set_xlabel('Вісь X')
axes_2.set_ylabel('Вісь Y')
axes_2.set_title('Парабола')

axes_3.plot(x, y, 'b')
axes_3.set_xlabel('Вісь X')
axes_3.set_ylabel('Вісь Y')
axes_3.set_title('Парабола')
```
